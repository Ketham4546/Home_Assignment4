```python
from transformers import pipeline

# Load the pre-trained sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Input sentence
sentence = "Despite the high price, the performance of the new MacBook is outstanding."

# Analyze the sentiment
result = sentiment_pipeline(sentence)[0]

# Extract and print the sentiment label and confidence score
print(f"Sentiment: {result['label']}")
print(f"Confidence Score: {result['score']:.4f}")
```

```python
#1. What is the main architectural difference between BERT and GPT? Which uses an encoder and which uses a decoder?
#Ans:\wer. BERT (Bidirectional Encoder Representations from Transformers) is based solely on the encoder architecture of the transformer model, enabling it to understand context from both directions (left and right) of a word. In contrast, GPT (Generative Pre-trained Transformer) is built using only the decoder portion of the transformer and processes text in a left-to-right (unidirectional) manner.
#BERT: Encoder-based → Best for understanding and classification tasks.
#GPT: Decoder-based → Best for generative tasks like text generation.
```

```python
#2. Why is using pre-trained models like BERT or GPT beneficial for NLP applications compared to training from scratch?

#Answer: Using pre-trained models provides several advantages:
#Reduced Training Time: Pre-trained models have already learned general language representations, so they require less data and time to fine-tune for specific tasks.
#Improved Accuracy: These models are trained on massive corpora and capture complex language patterns, leading to better performance across a wide range of NLP tasks.
#Resource Efficiency: Training a transformer model from scratch requires significant computational resources, which may not be practical. Pre-trained models offer an efficient starting point.
#Transfer Learning: They can be fine-tuned for various tasks (like sentiment analysis, question answering, or summarization) without needing task-specific architectures from scratch.
```
